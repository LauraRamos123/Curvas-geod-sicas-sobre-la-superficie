```
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
def sphere(r):
    u = np.linspace(0, 2 * np.pi, 200)
    v = np.linspace(0, np.pi, 200)
    d = (r) * np.outer(np.cos(u), np.sin(v))
    e = (r) * np.outer(np.sin(u), np.sin(v))
    g = (r) * np.outer(np.ones(np.size(u)), np.cos(v))
    return d,e,g
N=100
I=1000
h=10/(N-1)

t=np.zeros(N)
te=np.zeros(N) #theta
f=np.zeros(N) #phi

for n in range(N):
  f[n]=np.random.rand()
  te[n]=np.random.rand()
  


te[0]=np.pi/2 #punto de partida en theta
te[N-1]=np.pi/4 #punto de llegada en theta
f[0]=-np.pi/2 #punto de partida en fi 
f[N-1]=np.pi #punto de llegada en fi 
for k in range(N-1):
    t[k+1]=t[k]+h
    
for i in range(1,I):
    for j in range(1,N-1):
        te[j]=(2*(h**2)*np.sin(2*te[j-1])-np.cos(2*te[j-1])*te[j-1]*((f[j+1]-f[j-1])**2)-4*te[j+1]-4*te[j-1])/((-4*(h**2)*np.cos(2*(te[j-1])))-8)
        f[j]=((np.cos(te[j])/np.sin(te[j]))*(te[j+1]-te[j-1])*(f[j+1]-f[j-1])+2*f[j+1]+2*f[j-1])/4
    
#trasnformación a coordenadas triangulares
z=np.cos(te)
x=np.sin(te)*np.cos(f)
y=np.sin(te)*np.sin(f)
fig = plt.figure()
bx= fig.add_subplot(111,projection='3d')
bx.plot3D(x,y,z, 'blue')
d,e,g = sphere(1)
ax = fig.gca(projection='3d')
ax.plot_surface(d, e, g, rstride=3, cstride=3, color=(1,0.8,0,0.1))

print(f)
```
