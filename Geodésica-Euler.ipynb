```
# En este programa se resuelve la ecuación geodésica con el método de Euler

from math import sin, cos, tan
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import math

def funcionx (x,y,theta):
    return sin(theta)*cos(theta)*y**2

def funciony (x,y,theta):
    return -2*(1/tan(theta))*x*y
     
def funciontheta (x):
    return x

def funcionphi (y):
    return y

n=300   #int(input("Numero de iteraciones "))
h=math.pi/(2*(n-1))   #float(input("Tamaño del paso: ")) Verificar trayectoria en la solución
x = np.zeros(n)
y = np.zeros(n)
theta = np.zeros(n)
phi = np.zeros(n)
x[0]=-1/np.sqrt(2) #derivada de theta
y[0]=1/np.sqrt(2) #derivada de phi
theta[0]=math.pi/2
phi[0]=0

for i in np.arange(1,n):
    x[i]=x[i-1]+(funcionx(x[i-1],y[i-1],theta[i-1]))*h
    y[i]=y[i-1]+(funciony(x[i-1],y[i-1],theta[i-1]))*h
    theta[i]=theta[i-1]+(funciontheta(x[i-1]))*h
    phi[i]=phi[i-1]+(funcionphi(y[i-1]))*h
    phi[i]=phi[i]%(2*np.pi)

A = np.zeros(n)
B = np.zeros(n)
C = np.zeros(n)

for i in np.arange(0,n):
    A[i]=sin(theta[i])*cos(phi[i])
    B[i]=sin(theta[i])*sin(phi[i])
    C[i]=cos(theta[i])
    
#get_ipython().run_line_magic('matplotlib', 'inline')

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(A, B, C, 'blue')
plt.show()

plt.plot(theta,label="theta")
plt.plot(phi,label="phi")
plt.legend()
plt.show()

print("ángulo phi final",round(phi[-1],2))
print("ángulo theta final",round(theta[-1],2))


print(A[-1])
print(B[-1])
print(C[-1])
```
