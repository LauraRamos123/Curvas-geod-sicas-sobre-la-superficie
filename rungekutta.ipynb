```python
import numpy as np
import matplotlib.pyplot as plt
N=300
h=20/N
t=np.zeros(N)
y=np.zeros(N)
x=np.zeros(N)
y[0]=2
x[0]=1

def f(x,y):
    d=3*x-2*x*y
    return d

def g(x,y):
    d=-y+3*x*y
    return d


for i in  range(N-1):
    t[i+1]=t[i]+h
    kx1=f(x[i],y[i])
    ky1=g(x[i],y[i])
    kx2=f(x[i]+(kx1*h/2),y[i]+(ky1*h/2))
    ky2=g(x[i]+(kx1*h/2),y[i]+(ky1*h/2))
    kx3=f(x[i]+(kx2*h/2),y[i]+(ky2*h/2))
    ky3=g(x[i]+(kx2*h/2),y[i]+(ky2*h/2))
    kx4=f(x[i]+kx3*h,y[i]+ky3*h)
    ky4=g(x[i]+kx3*h,y[i]+ky3*h)
    x[i+1]=x[i]+(h/6)*(kx1+2*kx2+2*kx3+kx4)
    y[i+1]=y[i]+(h/6)*(ky1+2*ky2+2*ky3+ky4)
    
plt.plot(x,y)
plt.plot(t,x)
plt.plot(t,y)
```
